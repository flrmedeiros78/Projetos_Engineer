```python
%sql
SELECT 
  upper(cli.cliente) AS CLIENTE
 ,replace(substr(cli.data_compra,1,10),'-','') AS DATA_COMPRA
 ,cast(cli.preco_btc AS DECIMAL(10,2)) AS PRECO_BTC
 ,cast(cli.quantidade_btc AS DECIMAL(10,2)) AS QTD_BTC
 ,cast(cli.quantidade_btc AS DECIMAL(10,3)) AS QTD_BTC
FROM DEFAULT.transacoes_cliente cli
```
